<p style="text-align:justify;">La fonction <code>print()</code> permet d'afficher une chaîne de caractères ("string" en anglais).
La chaîne de caractère que l'on souhaite afficher doit être écrite entre guillemets (simples ' ou doubles "). Sans guillemets, c'est le contenu de la variable dont le nom est transmis en paramètre qui sera affiché.</p>

Une variable peut contenir une chaîne de caractères.

```python
message = "Hello"
print(message)
type(message)
```

On peut concaténer (fusionner) plusieurs chaînes de caractère au moyen de l'opérateur <code>+</code> :

```python
sujet = "Alice "
verbe = "aime "
cod = "les glaces"
point = "."
phrase = sujet + verbe + cod + point
print(phrase)
```

On ne peut pas concaténer une chaîne de caractères avec une valeur numérique directement. Il faut au préalable transformer la valeur numérique en chaîne de caractères au moyen de la fonction <code>str()</code> :

```python
age = 25
début = "J'ai "
fin = " ans."
phrase = début + str(age) + fin
print(phrase)
```

Les chaines de caractères formatées (aussi appelées f-strings) permettent d’inclure la valeur d’expressions Python dans des chaines de caractères en les préfixant avec f et en écrivant les expressions entre accolades.

```python
age = 21
print(f"J'ai {age} ans.")
```

<h2> Partie 4. Saisie d'une entrée par l'utilisateur </h2>

On peut demander à l'utilisateur de saisir une valeur avec la fonction <code>input()</code>. On peut ajouter en paramètre facultatif un message qui donne des précisions sur la valeur attendue.

```python
a = input()
b = input("Entrez votre âge : ")
```

La valeur renvoyée par la fonction <code>input()</code> est du type chaîne de caractères (string). Si on souhaite utiliser cette valeur pour effectuer des calculs, on doit d'abord convertir cette chaîne en entier avec la fonction <code>int()</code> ou en flottant avec la fonction <code>float()</code>.

```python
saisie = input("Veuillez entre votre âge : ")
age = int(saisie)
age = age + 5
print(f"Dans 5 ans, vous aurez {age} ans.")
```

Si on veut saisir des nombres définis par une expression algébrique comme $21+5$ ou $\sqrt{2}$ et les utiliser pour un calcul, les fonctions respectives <code>int()</code> ou <code>float()</code> ne fonctionnent pas.  

```python
saisie = input("Entrer une somme de deux entiers : ")
nombre = int(saisie)
print(f"La somme vaut {nombre}.")
```

```python
saisie = input("Entrer la racine carrée d'un entier : ")
nombre = float(saisie)
print(f"Une valeur approchée de cette racine est {nombre}.")
```

Pour saisir une expression algébrique et qu'elle soit interprétée comme un nombre, il faut utiliser la fonction <code>eval()</code>.

```python
saisie = input("Entrer une somme de deux entiers : ")
nombre = eval(saisie)
print(f"La somme vaut {nombre}.")
```

```python
saisie = input("Entrer la racine carrée d'un entier : ")
nombre = eval(saisie)
print(f"Une valeur approchée de cette racine est {nombre}.")
```
